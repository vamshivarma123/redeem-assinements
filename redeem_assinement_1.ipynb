```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.datasets import load_iris
iris_data = load_iris()

iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris["species"] = iris_data.target
iris["species"] = iris["species"].map({0: "setosa", 1: "versicolor", 2: "virginica"})

iris.head()
```

```python

print(iris.isnull().sum())
iris = iris.fillna(iris.mean(numeric_only=True))
```

```python

print("Sepal Length - Mean:", iris["sepal length (cm)"].mean())
print("Sepal Length - Median:", iris["sepal length (cm)"].median())
print("Sepal Length - Max:", iris["sepal length (cm)"].max())
print("Sepal Length - Min:", iris["sepal length (cm)"].min())

print("\nPetal Length - Mean:", iris["petal length (cm)"].mean())
print("Petal Length - Median:", iris["petal length (cm)"].median())
print("Petal Length - Max:", iris["petal length (cm)"].max())
print("Petal Length - Min:", iris["petal length (cm)"].min())
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=iris)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(iris["sepal length (cm)"], kde=True, bins=20, color="purple")
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="species", y="petal length (cm)", data=iris, estimator=np.mean)
plt.title("Average Petal Length by Species")
plt.show()
```
